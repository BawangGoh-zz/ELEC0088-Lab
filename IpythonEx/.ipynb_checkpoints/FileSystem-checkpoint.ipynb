<u>**Part 1)** </u>: <br>
Implement a program that reads the content of a Wireshark (https://www.wireshark.org) output file into memory (dictionaries). Run Wireshark in your computer and produce a CSV file. Alternatively download an example file from here https://www.ee.ucl.ac.uk/⇠mrio/ws.csv

```python
import csv
import os
import pandas as pd

# Getting the current work directory (cwd)
thisdir = os.getcwd()
print(thisdir)

# r=root, d=directories, f = files
for r, d, f in os.walk(thisdir):
    for file in f:
        if ".csv" in file:
            print(os.path.join(r, file))

with open('./ws.csv', 'r') as csvfile:
    readCSV = csv.reader(csvfile, delimiter = ',')
    csvDict = {}
    for row in readCSV:
        keys = row[0]
        csvDict[keys] = row[1:]
csvfile.close()
```

<u>**Part 2)** </u>: <br>
Write a python program that takes the dictionary in the previous exercise and prints the list of flows in the file. A ﬂow is a set of packets with the same IP source, IP destination, TCP source port and TCP destination port

```python
def printFlows(tcpDict):
    # Define the new dictionary for just the TCP packets
    new_Dict = {}
    for keys, value in tcpDict.items():
        if ('>' in value[5]):
            new_Dict[keys] = value
    return new_Dict

new_Dict = printFlows(csvDict)

# Construct dictionary with keys and the corresponding values of IP TCP and byte length
def sameFlows(tcpFlow):
    # Define the new dictionary for the same flows
    TCPIP = {}
    for key, value in tcpFlow.items():
        TCP = [int(i) for i in value[5].split() if i.isdigit()]

        # Group the TCP source and destination port and IP source and destination port into a key
        info = (value[1],value[2],TCP[0],TCP[1])
        byte = value[4]

        # Overwrite the exisiting dictionary value with new values
        if info in TCPIP:
            TCPIP[info].append(byte)
        else:
            TCPIP[info] = [byte]
#             print(TCPIP)
    return byte, TCP, TCPIP

# Construct new dictionary with similar IP and TCP port as keys and bytes length as values
def newDict(byteDict):
    # Sum the total byte length for each IP and TCP port
    byteSum = {key: sum(map(int, value)) for key, value in byteDict.items()}
    #byteSum = {key: sum(value) for key, value in byteDict.items()}
    return byteSum
```

```python
# Debug purpose
info, test, flow = sameFlows(new_Dict)  
totalByte = newDict(flow)
# print(totalByte)

# Convert dictionary to pandas Dataframe
# pd.DataFrame(totalByte.values(), columns=['IP source', 'IP destination', 'TCP source', 'TCP destination'])
pd.DataFrame([[keys[0], keys[1], keys[2], keys[3], values] for keys, values in totalByte.items()]) \
            .rename(columns={0:'IP source', 1:'IP destination', 2:'TCP source', 3:'TCP destination', 4:'Byte length'})
```
